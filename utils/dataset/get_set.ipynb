```python
import os
import shutil

# Define the paths
input_file = 'filelists/test_filelists_wa.txt'
source_folder = 'wallon_dataset - sentences'
destination_folder = 'wallon_test_set'

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Read the text file and extract the file names
with open(input_file, 'r', encoding='utf-8') as file:
    lines = file.readlines()

for line in lines:
    # Extract the file name
    file_name = line.split('/')[1].split('|')[0].strip()
    
    # Define the full path to the source and destination files
    source_file = os.path.join(source_folder, file_name)
    destination_file = os.path.join(destination_folder, file_name)
    
    # Copy the file to the destination folder
    if os.path.exists(source_file):
        shutil.copy(source_file, destination_file)
        print(f"Copied: {source_file} to {destination_file}")
    else:
        print(f"File not found: {source_file}")
```
